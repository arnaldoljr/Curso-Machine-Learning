## Artificial Neural Network



```python
# Artificial Neural Network

# Installing Theano
# pip install --upgrade --no-deps git+git://github.com/Theano/Theano.git

# Installing Tensorflow
# Install Tensorflow from the website: https://www.tensorflow.org/versions/r0.12/get_started/os_setup.html

# Installing Keras
# pip install --upgrade keras

# Part 1 - Data Preprocessing
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Churn_Modelling.csv')
dataset
```

```python
X = dataset.iloc[:, 3:13].values
X
```

```python
y = dataset.iloc[:, 13].values
y
```

Criando as variáveis categóricas para os países e os gêneros

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
X
```

```python
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
X

```

É preciso informar que sao variáveis categóricas, ou seja, 0,1,2 nao indica um peso e sim uma  categoria

( criar as variáveis dummy ) - criar tem colunas no início para representar os países - tipo 0 - 1 0 0 ,1 - 0 1 0 e o 2 - 0 0 1 - a biblioteca adiciona estas variáveis antes das outras e remove o campo característico dos países

```python
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:] # remove a primeira coluna pq nao preciso
X
```

```python
X[0,:]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
X_train
```

```python
X_test
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
# Part 2 - Now let's make the ANN!

# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense

```

```python
# Initialising the ANN
classifier = Sequential()
```

OBS- #ReLU
A função ReLU é a unidade linear rectificada. É definida como:

f(x) = max (0, x)

Não ativa todos os neurônios ao mesmo tempo. O que isto significa ? Se você olhar para a função ReLU e a entrada for negativa, ela será convertida em zero e o neurônio não será ativado. Isso significa que, ao mesmo tempo, apenas alguns neurônios são ativados, tornando a rede esparsa e eficiente e fácil para a computação.


```python
# Adding the input layer and the first hidden layer
#init = 'uniform' indica que os pesos serão iniciados com uma distribuição uniforme
# tipo de função de ativaçao de cada neurônio - activation = 'relu'
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 11))
```

```python
# Adding the second hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))
```

```python
# Adding the output layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
```

No caso de uma camada de saída com 3 categorias deve-se mundar a dimensão de output e a função de ativação seria a softmax 

```python
classifier
```

optimizer = 'adam'-> indica que o tipo de gradiente descendente estocástico é o 'adam' 

loss = 'binary_crossentropy' -> indica que a função de erro será a função logaritmica para saída binária  

metrics = ['accuracy'] -> é o modo como ele atualiza os pesos da rede neural

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

```

```python
classifier
```

batch_size = 10 -> indica que após 10 observações que os pesos serão atualizados

nb_epoch = 100 -> todos os dados passarão 100 vezes pela rede no treinamento



```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
#probabilidada de 23% de deixar o banco
y_pred[0]
```

```python
y_pred = (y_pred > 0.5)
y_pred
```

```python
#indica que o cliente zero nao deixou o banco
y_pred[0]
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
