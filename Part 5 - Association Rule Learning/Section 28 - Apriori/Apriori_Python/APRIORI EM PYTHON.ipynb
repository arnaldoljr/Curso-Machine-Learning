```python
# preciso tirar os cabeçalhos
# Data Preprocessing
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
```

```python
dataset
```

```python
#transformar os dados em uma lista de listas
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

#### parâmetros 

min_support - indica o mínimo de vezes que devem aparecer nos dados; no caso ele quer que aparecem no mínimo 3 vezes por semana

```python
3*7/7500
```

min_confiance indica a quantidade de vezes que a regra de estar certa, ou seja, o valor de 80% indica que a regra deve estar certa em 80% das vezes que os parâmetros aparecerem

```python
# Training Apriori on the dataset
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
# Visualising the results
results = list(rules)
```
