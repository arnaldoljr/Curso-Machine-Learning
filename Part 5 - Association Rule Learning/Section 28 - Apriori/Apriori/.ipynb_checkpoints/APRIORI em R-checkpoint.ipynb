```r
# Apriori

# Data Preprocessing
install.packages('arules')
```

```r
library(arules)
```

```r
#ler os dados
dataset = read.csv('Market_Basket_Optimisation.csv', header = FALSE)
dataset
```

```r
#transformar cada nome de instância em um valor associado
#removendo as duplicadas
dataset = read.transactions('Market_Basket_Optimisation.csv', sep = ',', rm.duplicates = TRUE)
```

Há 1 duplicata ocorrendo 5 vezes nos dados.

```r
#analisar como está a matriz esparsa 
summary(dataset)
```

CRIAÇÃO DAS REGRAS 
Coloca um support de um frequência de 3 elementos em uma semana - temos um total de 7500. Ou seja, ele quer pegar todo mundo que tenha aparecido pelo menos 21 vezes ( 3 vezes por dia durante uma semana) 

```r
3*7/7500 
```

A confiança indica qual a frenquência que tem que aparecer nos dados. Ele inicia os testes com uma confiança de 80%. Ou seja, os elementos que compoem as regras precisam aparecer em 80% dos dados( regras ).

```r
#colocar uma confiaça de 
# Training Apriori on the dataset
rules = apriori(data = dataset, parameter = list(support = 0.003, confidence = 0.8))
```

NENHUMA REGRA FOI ENCONTRADA. DIMINUI PARA 40% DE FREQUÊNCIA.

```r
# Training Apriori on the dataset
rules = apriori(data = dataset, parameter = list(support = 0.003, confidence = 0.4))
```

ELE ENCONTRA 281 REGRAS

```r
# Visualising the results
# pega as 10 melhores regras ordenadas pelas melhores 'lifts'

inspect(sort(rules, by = 'lift')[1:10])
```

Interpretaçao - >  [2]  {spaghetti,tomato sauce}                    => {ground beef}      
as pessoas que compram spaghetti,tomato sauce, comprar também ground beef 48% das vezes ([2]  0.003066258 0.4893617  4.980600 23   
)

```r
# Training Apriori on the dataset
#Colocando a confiança em 20%
rules = apriori(data = dataset, parameter = list(support = 0.003, confidence = 0.2))

```

TEMOS AGORA 1348 Regras

```r
# Visualising the results
# pega as 10 melhores regras ordenadas pelas melhores 'lifts'

inspect(sort(rules, by = 'lift')[1:10])
```
